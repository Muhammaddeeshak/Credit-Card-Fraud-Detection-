# Histogram showing Distribution each variable 

```python
fig, axs = plt.subplots(6, 5, figsize=(15, 15))

x_columns = [
    'V1', 'V2', 'V3', 'V4', 'V5', 'V6',
    'V7', 'V8', 'V9', 'V10', 'V11',
    'V12', 'V13', 'V14', 'V15', 'V16',
    'V17', 'V18', 'V19', 'V20', 'V21',
    'V22', 'V23', 'V24', 'V25', 'V26',
    'V27', 'V28', 'Time', 'Amount'
]

for i, x_column in enumerate(x_columns):
    row = i // 5
    col = i % 5

    axs[row, col].hist(credit_card[x_column], bins=20, color='skyblue', edgecolor='black')
    axs[row, col].set_xlabel(x_column)
    axs[row, col].set_ylabel('Frequency')
    axs[row, col].set_title(f'Distribution of {x_column}')

plt.tight_layout()
plt.show()
```

# Scatterplot Showing relationships b/w Variables 

```python

# Create a figure with a 6x5 grid of subplots
fig, axs = plt.subplots(6, 5, figsize=(15, 15))

# Define the data you want to plot
x = credit_card['V1']
y_columns = [
    'V2', 'V3', 'V4', 'V5', 'V6',
    'V7', 'V8', 'V9', 'V10', 'V11',
    'V12', 'V13', 'V14', 'V15', 'V16',
    'V17', 'V18', 'V19', 'V20', 'V21',
    'V22', 'V23', 'V24', 'V25', 'V26',
    'V27', 'V28', 'Time', 'Amount'
]

# Loop through the columns and create scatter plots
for i, y_column in enumerate(y_columns):
    row = i // 5
    col = i % 5

    axs[row, col].scatter(x, credit_card[y_column], label='Data point', color='blue', marker='.')
    axs[row, col].set_xlabel('V1')
    axs[row, col].set_ylabel(y_column)
    axs[row, col].set_title(f'V1 vs {y_column}')
    axs[row, col].legend()

# Adjust layout and display the figure
plt.tight_layout()
plt.show()
```

Using pipeline to make preprocesssing in squential form by using PCA for further reducing demensions of data into 5 independent Variables.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

pipe_dt=Pipeline([('Scaler',StandardScaler()),('PCA',PCA(n_components=5))])

pipe_dt.fit(credit_card)
credit_card_scaled=pipe_dt.transform(credit_card)
```

```python


import pandas as pd

# If your data is in a NumPy array, you can convert it to a DataFrame
# Assuming credit_card_scaled is a NumPy array
credit_card_scaled_df = pd.DataFrame(credit_card_scaled)

# Now, you can easily view it in tabular form
credit_card_scaled_df

```

```python
mean = np.mean(credit_card_scaled_df)
std = np.std(credit_card_scaled_df)
print('Mean:', mean)
print('Standard deviation:', std)
```

# Splited data into train and test with Train = 80% and test = 20%

```python
from sklearn.model_selection import train_test_split, GridSearchCV

f_train,f_test,t_train,t_test = train_test_split(credit_card_scaled_df,y,test_size= 0.2, random_state=12)
```

# Support Vector Machine Classifier Uses 'c','kernel' and 'gamma' as their Hypermaters and GridSearchCv run SVC with it's different combinations of hyperparameters for best combination, which maximizes accuracy measures . 

```python
from sklearn import svm 
```

```python
clf=svm.SVC()
```

```python
param_grid=[{'C':[0.1,1.0,10.0],'kernel':['linear','rbf'],'gamma':['scale','auto']}]

grid_search= GridSearchCV(clf,param_grid,cv=5,scoring='accuracy',return_train_score=True, verbose=10)

grid_search.fit(f_train,t_train)
```

```python
clf_predict=grid_search.predict(f_test)
```

```python
final_clf=grid_search.best_estimator_
final_clf
```

```python
final_clf.score(f_test,t_test)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score 
print(confusion_matrix(t_test,clf_predict))
print(precision_score(t_test,clf_predict))
print(recall_score(t_test,clf_predict))
print(f1_score(t_test,clf_predict))
```
